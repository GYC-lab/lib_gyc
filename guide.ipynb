# plot data with gplot()

## generate data

```python
# ------------------------------- generate data ------------------------------ #
x = np.linspace(0, 1, 100)
y_1 = x**(1/2)
y_2 = x**(1/3)
y_3 = x**(1/4)
y_4 = x**(1/5)
```

## plot data

```python
# ---------------------------- create a new Figure --------------------------- #
figname = 'my_plot'
figsize = (6,4.5)
plt.figure(figname,figsize=figsize)
# -------------------------------- first curve ------------------------------- #
settings                  = gplot_style()   # get default settings
settings['line']["label"] = '$y=x^{1/2}$'   # add label
settings['line']['color'] = 'k'             # change color
settings['line']["style"] = '-'             # change linestyle
gplot(x,y_1,settings)                       # plot
# ------------------------------- second curve ------------------------------- #
settings                      = gplot_style()       # get default settings
settings['line']["label"]     = '$y=x^{1/3}$'       # add label
settings['line']['color']     = '#D20000'           # change color
settings['line']["linestyle"] = ''                 # change linestyle
settings['line']["marker"]    = 's'                 # add marker
settings['line']["markevery"] = 10                  # change marker frequency
gplot(x,y_2,settings)
# -------------------------------- third curve ------------------------------- #
settings                      = gplot_style()     # get default settings
settings['line']["label"]     = '$y=x^{1/4}$'       # add label
settings['line']['color']     = '#2d2dff'           # change color
settings['line']["style"]     = '-.'             # change linestyle
settings['line']["marker"]    = '+'                 # add marker
settings['line']["markevery"] = 10                  # change marker frequency
settings['line']["linestyle"] = '--'                # change linestyle
gplot(x,y_3,settings)
# ------------------------------- fourth curve ------------------------------- #
settings                      = gplot_style()   # get default settings
settings['line']["label"]     = '$y=x^{1/5}$'   # add label
settings['line']['color']     = '#00D200'       # change color
settings['line']["style"]     = ':'             # change linestyle
settings['line']["marker"]    = 'x'             # add marker
settings['line']["markevery"] = 10              # change marker frequency
settings['line']["linestyle"] = '-.'            # change linestyle
# ----------------------- settings for the whole figure ---------------------- #
settings['axis']["xmin"]    = 0                 # change min of x
settings['axis']["xmax"]    = 1                 # change max of x
settings['title']['name']   = '$y=x^{1/n}$'     # add title
settings['leg']['location'] = 'lower right'     # change legend location
gplot(x,y_4,settings)
# --------------------------------- save fig --------------------------------- #
# default format is png, set _format='svg' or _format='pdf' to save as svg or pdf
gsavefig(figname)
# plt.close(figname)
```

# plot data with gcontour()

## generate data

```python
n_points = 200
x_1d     = np.linspace(-4, 4, n_points)
y_1d     = np.linspace(-4, 4, n_points)
x_2d, y_2d = np.meshgrid(x_1d, y_1d)
z_2d = np.cos(x_2d)*np.cos(y_2d)
```

## plot data

```python
# ---------------------------- create a new Figure --------------------------- #
figname = 'my_contour'
figsize = (6,4.5)
plt.figure(figname,figsize=figsize)

# ------------------------------- plot contour ------------------------------- #
settings                    = gcontour_style()      # get default settings
settings["title"]           = '$u=\cos(x)\cos(y)$'  # add title
settings["xlabel"]          = '$x$'                 # add xlabel
settings["ylabel"]          = '$y$'                 # add ylabel
settings['yrotation']       = 0                     # change ytick rotation
settings["vmin"]            = -1                    # change min of colorbar
settings["vmax"]            = 1                     # change max of colorbar
settings["cmap"]            = cmaps.MPL_RdBu_r      # change colormap
settings["shading"]         = 'gouraud'             # change shading
settings["alpha"]           = 1                     # change alpha to make contour transparent
settings["show_colorbar"]   = True                  # show colorbar
settings["bar_label"]       = '$u$'                 # add colorbar label
settings["bar_tick_length"] = 12                    # change colorbar tick length
settings["bar_num_ticks"]   = 5                     # change colorbar number of ticks
settings["bar_position_size"] = [0.85, 0.15, 0.05, 0.7] # change colorbar position and size
gcontour(x_2d, y_2d, z_2d, settings)                # plot

# ----------------------------- plot contourline ----------------------------- #
settings               = gcontourline_style()   # get default settings
settings["levels"]     = [0]                    # change contour levels
settings["linestyles"] = ['--']                 # change contour line style
settings["linewidths"] = [2]                    # change contour line width
settings["colors"]     = ['k']                  # change contour line color
gcontourline(x_2d, y_2d, z_2d,settings)          # plot

# --------------------------------- save fig --------------------------------- #
# default format is png, set _format='svg' or _format='pdf' to save as svg or pdf
gsavefig(figname)
```
